```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Импортируем вычищенный в Excel датасет
dataset = pd.read_excel('Chisty_dataset.xlsx')
dataset = dataset.loc[:155,:]
#Переводим все форматы к необходимому
dataset['Пропустил уроков'] = dataset['Пропустил уроков'].apply(lambda x : int(x))
dataset['Посетил уроков'] = dataset['Посетил уроков'].apply(lambda x : int(x))
dataset.replace(['NaN', 'NaT'], np.nan, inplace = True)
#удаляем строчки с пропущенными данными
dataset = dataset.dropna(subset = ['Возраст'])
#берем только курс 8-12
dataset = dataset[dataset['Тип курса'] == '2019-12-08']
#Анализируем только тех, кто хотя бы раз пришел
dataset = dataset[dataset['Посетил уроков']!=0]
dataset
```

```python
#Посмотрим, какие данные у нас есть по посещаемости
dlin = dataset['Посещаемость'].str.len().value_counts()
dlin
```

```python
#Удалим записи о тех, кто еще в процессе прохождения курса
modified = dataset[dataset['Посещаемость'].str.len()>29]
#Оставим только записи о посещенных первых 15 занятиях, чтобы понять, как изменяется вероятность того, что ученик будет отчислен
modified['Посещаемость'] = modified['Посещаемость'].apply(lambda x: x[:29])

_, axes = plt.subplots(1, 2, sharey=True, figsize=(17, 8))

sns.countplot(x='Пропустил уроков', hue='Отчислен.1', data=modified, ax=axes[0])
sns.countplot(x='Посетил уроков', hue='Отчислен.1', data=modified, ax=axes[1]);
```

```python
modified['Посещаемость'] = modified['Посещаемость'].apply(lambda x: np.fromstring(x.replace(',',','), sep=','))
modified.head()
```

```python
expelled = modified[modified['Отчислен.1'] == 1]
finished = modified[modified['Отчислен.1'] == 0]
probability_exp=expelled['Посещаемость'].sum(axis=0)/21
probability_fin=finished['Посещаемость'].sum(axis=0)/81
pe=pd.DataFrame(probability_exp)
pf=pd.DataFrame(probability_fin)
```

```python
from scipy.sparse import csr_matrix, hstack
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score
from sklearn.linear_model import LogisticRegression,SGDClassifier
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import validation_curve, learning_curve

sns.set()

X_train, X_test, y_train, y_test = train_test_split(modified['Посещаемость'], 
                                                    modified['Отчислен.1'], test_size=0.2, random_state=42)
y_train = y_train.as_matrix()
y_test = y_test.as_matrix()
X_test = np.stack(X_test,axis=1)
X_train = np.stack(X_train,axis=1)
X_train = X_train.reshape((81,15))
X_test = X_test.reshape((21,15))
```

```python
logit = LogisticRegression(C=2,random_state=17)
#regularization
cv_scores=cross_val_score(logit,X_train,y=y_train, cv=4,scoring='roc_auc')
```

```python
alphas = np.logspace(-2, 0, 20)
sgd_logit = SGDClassifier(loss='log', n_jobs=-1, random_state=17, max_iter=5)
logit_pipe = Pipeline([('scaler', StandardScaler()), ('poly', PolynomialFeatures(degree=2)), 
                       ('sgd_logit', sgd_logit)])
val_train, val_test = validation_curve(logit_pipe, X_train, y_train,
                                       'sgd_logit__alpha', alphas, cv=5,
                                       scoring='roc_auc') 
```

```python
def plot_with_err(x, data, **kwargs):
    mu, std = data.mean(1), data.std(1)
    lines = plt.plot(x, mu, '-', **kwargs)
    plt.fill_between(x, mu - std, mu + std, edgecolor='none',
                     facecolor=lines[0].get_color(), alpha=0.2)

plot_with_err(alphas, val_train, label='training scores')
plot_with_err(alphas, val_test, label='validation scores')
plt.xlabel(r'$\alpha$'); plt.ylabel('ROC AUC')
plt.legend()
plt.grid(True);
```

```python
l1 = logit_pipe.fit(X_train,y_train)
results = logit_pipe.predict(X_test)
```

```python
from sklearn.metrics import roc_curve, auc
fpr, tpr, thresholds = roc_curve(y_test, results)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=1, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=1, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
```
